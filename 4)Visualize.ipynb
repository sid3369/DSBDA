```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
data = pd.read_csv("heart.csv")
```

```python
data
```

```python

data["age"] = data["age"].fillna(data["age"].mean())
```

```python

# Pie Chart
plt.pie([data["sex"].mean(),data["trestbps"].mean(),data["chol"].mean(),data["slope"].mean()])
plt.plot()
```

```python
# Histogram
plt.hist(x=data["age"], bins=20, color="orange", label="Age")
plt.title("age")
plt.xlabel("age")
plt.ylabel("chol")
```

```python
# Linegraph
plt.plot(data.iloc[0:30, 6], data.iloc[0:30, 3], color="red")
plt.title("RestECG vs AHD")
plt.xlabel("RestECG")
plt.ylabel("AHD")
```

```python
# Scatter Plot
plt.scatter(data.iloc[0:30, 6], data.iloc[0:30, 3], color="red")
plt.title("Age vs Chol")
plt.xlabel("Age")
plt.ylabel("Chol")
```
